Processing the frames of a video file

```python
import cv2
import matplotlib.pyplot as plt
```

```python
capt = cv2.VideoCapture('globe.mp4')
```

```python
i=0
while(capt.isOpened()):
  r,f = capt.read()
  if r==False:
    break
  cv2.imwrite('abc'+str(i)+'.jpg',f)
  i=i+1
```

```python
numb_frames = i
```

```python
start_f = (int(numb_frames)*1/3)
stop_f = (int(numb_frames)*2/3)
```

```python
start_f=int(start_f)
start_f
```

```python
stop_f=int(stop_f)
stop_f
```

```python
img=cv2.imread('abc0.jpg',0)
```

```python
sh=img.shape
sh
```

```python
strt_rw = int(sh[0]*3/10)
stp_rw = int(sh[0]*6/10)
strt_col = int(sh[1]*8/10)
stp_col = int(sh[1]*9/10)
print(strt_rw,stp_rw,strt_col,stp_col)
```

```python
for j in range(start_f,stop_f):
  img=cv2.imread('abc'+str(j)+'.jpg',0) # 0 means black and white
  img[strt_rw:stp_rw,strt_col:stp_col]=255
  cv2.imwrite('abc'+str(j)+'.jpg',img)
capt.release()
cv2.destroyAllWindows()
```
