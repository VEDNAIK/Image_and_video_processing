```python
image3=Image.open('peppers.png')
```

imread will convert the image to matrix.  
imshow will convert matrix to image on screen.    
Image is opening the image(PIL library)

```python
plt.figure()
plt.subplot(1,2,1) #if we do 2,1 then it will show column wise plt.imshow(image3)
plt.imshow(image3)
plt.title('png')
image3.save('peppers.jpg')
image4=Image.open('peppers.jpg')
plt.subplot(1,2,2)
plt.imshow(image4)
plt.title('jpg')
```

```python
image5 = imread('ishihara.jpg')
image5_gray = color.rgb2gray(image5)
plt.figure(figsize=(10,10))
plt.subplot(1,2,1)
plt.imshow(image5)
plt.subplot(1,2,2)
plt.imshow(image5_gray,cmap='gray')
```
