Detection of horizontal, vertical and diagonal edges using sobel filters.

```python
from skimage import io
import numpy as np
from scipy import signal
from skimage.color import rgb2gray
import matplotlib.pyplot as plt
```

```python
img = io.imread('roof.tif')
plt.imshow(img, cmap='gray')
```

# Horizontal Edges

```python
horizontal_filter = [[-1,-2,-1],[0,0,0],[1,2,1]]
img_filtered_h = signal.convolve2d(img, horizontal_filter, mode='same')# we write mode as after convolution
#image size is reduced so we want same image size so we use mode='same'
```

```python
plt.imshow(img_filtered_h, cmap = "gray")
```

```python
rows,cols = img.shape
for r in range(rows):
    for c in range(cols):
        if img_filtered_h[r,c]<0:
            img_filtered_h[r,c]=0
            # here eg the range is -10 to 100 and when we use imshow python converts it to positve range i.e. 0 to 255 but
            # the max value was 100 we want it as it is so we are converting -ve values to 0
```

```python
plt.imshow(img_filtered_h, cmap = "gray")
```

# Vertical edges

```python
vert_filter = [[-1,0,1],[-2,0,2],[-1,0,1]]
img_filtered_v = signal.convolve2d(img, vert_filter, mode='same')
```

```python
plt.imshow(img_filtered_v, cmap = "gray")
```

```python
rows,cols = img.shape
for r in range(rows):
    for c in range(cols):
        if img_filtered_v[r,c]<0:
            img_filtered_v[r,c]=0
```

```python
plt.imshow(img_filtered_v, cmap = "gray")
```

# Forward Diagonal

```python
forward_diagonal_filter = [[0,-1,-2],[1,0,-1],[2,1,0]]
img_filtered_fd = signal.convolve2d(img, forward_diagonal_filter, mode='same')
```

```python
rows,cols = img.shape
for r in range(rows):
    for c in range(cols):
        if img_filtered_fd[r,c]<0:
            img_filtered_fd[r,c]=0
```

```python
plt.imshow(img_filtered_fd, cmap = "gray")
```

# Backward Diagonal

```python
backward_diagonal_filter = [[0,1,2],[-1,0,1],[-2,-1,0]]
img_filtered_bd = signal.convolve2d(img, backward_diagonal_filter, mode='same')
```

```python
rows,cols = img.shape
for r in range(rows):
    for c in range(cols):
        if img_filtered_bd[r,c]<0:
            img_filtered_bd[r,c]=0
```

```python
plt.imshow(img_filtered_bd, cmap = "gray")
```

# All Edges

```python
img_filtered_all = img_filtered_h + img_filtered_v + img_filtered_fd + img_filtered_bd
plt.imshow(img_filtered_all, cmap = "gray")
```

```python
plt.figure(figsize=(30,20))

plt.subplot(2,3,1)
plt.imshow(img_filtered_h, cmap = "gray")
plt.title("Horizontal edges")
plt.subplot(2,3,2)
plt.imshow(img_filtered_v, cmap = "gray")
plt.title("Vertical edges")
plt.subplot(2,3,3)
plt.imshow(img_filtered_fd, cmap = "gray")
plt.title("Forward diagonal")
plt.subplot(2,3,4)
plt.imshow(img_filtered_bd, cmap = "gray")
plt.title("Backward diagonal")
plt.subplot(2,3,5)
plt.imshow(img_filtered_all, cmap = "gray")
plt.title("All edges")
plt.subplot(2,3,6)
plt.imshow(img, cmap = "gray")
plt.title("Orignal image")
```

# Laplacian Filter

```python
# It is a vraiant of laplacian filter as the matrix was 111 1-81 111 it takes 
# diagonals also orignal was 010 1-81 010
```

```python
# High pass filter
image = io.imread("fruits.png")
image = rgb2gray(image)
hp_filter = [[1,1,1],[1,-8,1],[1,1,1]]
img_filtered_hp = signal.convolve2d(image, hp_filter, mode='same')
```

```python
plt.figure(figsize=(8,8))
plt.imshow(img_filtered_hp, cmap = "gray")
```

```python
rows,cols = image.shape
for r in range(rows):
    for c in range(cols):
        if img_filtered_hp[r,c]<0:
            img_filtered_hp[r,c]=0
```

```python
plt.figure(figsize=(8,8))
plt.imshow(img_filtered_hp, cmap = "gray")
```

```python
image_edges = image + img_filtered_hp 
```

```python
plt.imshow(image_edges, cmap = "gray")
```

```python
plt.imshow(image, cmap = "gray")
```

```python
plt.figure(figsize=(8,8))
plt.subplot(1,2,1)
plt.imshow(image, cmap = "gray")
plt.title('Orignal image')
plt.subplot(1,2,2)
plt.imshow(image_edges, cmap = "gray")
plt.title('High pass filtered image (edge detection)')
```

Horizontal, vertical and diagonal filters are applied to highlight diagonal edges.  
To highlight all types of edges, all filtered images have been added.  
It is observed that all edges were highlighted.  
To improve the contrast of all the edges of the given image, all the negative values were converted to zero.  
Variant of laplacian is applied to highlight all types of egdes and filtered image shows that all types of edges are highlighted.
